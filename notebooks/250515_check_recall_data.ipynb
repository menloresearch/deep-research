```python
# Outer Structure (Python Dictionary representation):

{
    "env": "[See Python Code Below]",
    "func_schemas": "[See Function Schema JSON Below]",
    "index": "101cf927-7426-40ec-bf63-90525c194edd",
}

# Mock library management system environment

# Content of 'env' (Python Code):

from datetime import datetime

books = [
    {
        "id": "BK001",
        "title": "The Quantum Paradox",
        "author": "Alice Smith",
        "year": 2023,
        "copies": 3,
        "reserved_by": [],
        "due_date": None,
    },
    {
        "id": "BK002",
        "title": "Neural Networks Simplified",
        "author": "Bob Johnson",
        "year": 2021,
        "copies": 2,
        "reserved_by": ["USER456"],
        "due_date": datetime(2023, 11, 15),
    },
]

users = {
    "USER123": {"full_name": "John Doe", "overdue_books": 1},
    "USER456": {"full_name": "Jane Smith", "overdue_books": 0},
}


def search_books(title: str = None, author: str = None) -> list:
    """Search books by title or author"""
    results = []
    for book in books:
        if (title and title.lower() in book["title"].lower()) or (
            author and author.lower() in book["author"].lower()
        ):
            results.append(book)
    return results


def check_availability(book_id: str) -> dict:
    """Check available copies and existing reservations"""
    for book in books:
        if book["id"] == book_id:
            available = book["copies"] - len(book["reserved_by"])
            return {
                "available": available,
                "total_copies": book["copies"],
                "reserved_users": book["reserved_by"],
            }
    return {"error": "Book not found"}


def reserve_book(user_id: str, book_id: str) -> str:
    """Reserve a book for a user if available"""
    book = next((b for b in books if b["id"] == book_id), None)
    if not book:
        return "Book not found"
    if len(book["reserved_by"]) >= book["copies"]:
        return "No available copies"
    if user_id in book["reserved_by"]:
        return "Already reserved by user"

    book["reserved_by"].append(user_id)
    return f"Reservation successful. ID: RSV-{book_id}-{len(book['reserved_by'])}"


def calculate_late_fee(user_id: str) -> float:
    """Calculate overdue fees for a user"""
    user = users.get(user_id)
    if not user:
        return 0.0
    return user["overdue_books"] * 5.0  # $5 per overdue book


def pay_late_fee(user_id: str, amount: float) -> bool:
    """Process late fee payment and update user status"""
    if user_id not in users:
        return False
    if amount >= calculate_late_fee(user_id):
        users[user_id]["overdue_books"] = 0
        return True
    return False


# Content of 'func_schemas' (Function Schema JSON):

[
    {
        "type": "function",
        "function": {
            "name": "search_books",
            "description": "Search books by title or author",
            "parameters": {
                "type": "object",
                "properties": {
                    "title": {
                        "type": "string",
                        "description": "Title of the book to search for",
                    },
                    "author": {
                        "type": "string",
                        "description": "Author of the book to search for",
                    },
                },
                "required": [],
            },
        },
    },
    {
        "type": "function",
        "function": {
            "name": "check_availability",
            "description": "Check available copies and existing reservations",
            "parameters": {
                "type": "object",
                "properties": {
                    "book_id": {
                        "type": "string",
                        "description": "ID of the book to check availability for",
                    }
                },
                "required": ["book_id"],
            },
        },
    },
    {
        "type": "function",
        "function": {
            "name": "reserve_book",
            "description": "Reserve a book for a user if available",
            "parameters": {
                "type": "object",
                "properties": {
                    "user_id": {
                        "type": "string",
                        "description": "ID of the user reserving the book",
                    },
                    "book_id": {
                        "type": "string",
                        "description": "ID of the book to reserve",
                    },
                },
                "required": ["user_id", "book_id"],
            },
        },
    },
    {
        "type": "function",
        "function": {
            "name": "calculate_late_fee",
            "description": "Calculate overdue fees for a user",
            "parameters": {
                "type": "object",
                "properties": {
                    "user_id": {
                        "type": "string",
                        "description": "ID of the user to calculate late fees for",
                    }
                },
                "required": ["user_id"],
            },
        },
    },
    {
        "type": "function",
        "function": {
            "name": "pay_late_fee",
            "description": "Process late fee payment and update user status",
            "parameters": {
                "type": "object",
                "properties": {
                    "user_id": {
                        "type": "string",
                        "description": "ID of the user making the payment",
                    },
                    "amount": {
                        "type": "number",
                        "description": "Amount being paid towards late fees",
                    },
                },
                "required": ["user_id", "amount"],
            },
        },
    },
]
```
